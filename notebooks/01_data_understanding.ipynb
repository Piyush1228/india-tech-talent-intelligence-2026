```python
missing_count = df.isnull().sum()
missing_pct = (df.isnull().sum() / len(df) * 100).round(2)
missing_report = pd.DataFrame({'Missing_Count': missing_count,'Missing_Percentage': missing_pct}).sort_values('Missing_Count', ascending=False)
missing_report = missing_report[missing_report['Missing_Count']>0]
```

```python
# Are these the same 571 rows across all 5 columns?
cols_to_check = ['minimumSalary', 'maximumSalary', 'minimumExperience', 'maximumExperience', 'tagsAndSkills']
overlap = df[cols_to_check].isnull().all(axis=1).sum()
print(f"Rows missing ALL 5 of these columns simultaneously: {overlap}")
```

```python
categorical_cols = ['title', 'companyName', 'location', 'experience', 'currency']
```

```python
print("CARDINALITY CHECK")
print("=" * 60)
for col in categorical_cols:
    unique_count = df[col].nunique()
    total_count = len(df)
    print(f"{col:20s} | Unique: {unique_count:6d} | % Unique: {unique_count/total_count*100:.2f}%")
```

```python
skill_counts = df['tagsAndSkills'].dropna().apply(lambda x: len(x.split(',')))
print(f"Average skills per posting: {skill_counts.mean():.2f}")
print(f"Median skills per posting: {skill_counts.median()}")
print(f"Max skills in one posting: {skill_counts.max()}")
print(f"Min skills in one posting: {skill_counts.min()}")

# Check for any unusual delimiters (just comma, or mixed?)
sample_check = df['tagsAndSkills'].dropna().sample(20, random_state=42)
for s in sample_check:
    if '|' in s or '/' in s:
        print(f"NON-COMMA DELIMITER FOUND: {s[:80]}")
```

```python
print("LOGICAL CONSISTENCY CHECKS")
print("=" * 60)
salary_issue = df[df['minimumSalary'] > df['maximumSalary']]
print(f"Rows where min salary > max salary: {len(salary_issue)}")

exp_issue = df[df['minimumExperience'] > df['maximumExperience']]
print(f"Rows where min experience > max experience: {len(exp_issue)}")

dupe_jobs = df['jobId'].duplicated().sum()
print(f"Duplicate jobId entries: {dupe_jobs}")

full_dupes = df.duplicated().sum()
print(f"Fully duplicate rows: {full_dupes}")
```

```python
df['location'].value_counts().tail(30)
```

```python
# Step 1: Get all rows involved in a jobId duplicate (both copies)
jobid_dupes = df[df['jobId'].duplicated(keep=False)].copy()

# Step 2: Within THIS subset, check which rows are also FULL duplicates
jobid_dupes['is_full_dupe'] = jobid_dupes.duplicated(keep=False)

# Step 3: Partial duplicates = jobId matches, but full row does NOT match
partial_dupes = jobid_dupes[jobid_dupes['is_full_dupe'] == False]

print(f"Partial duplicates found: {len(partial_dupes)}")
print(partial_dupes[['jobId', 'title', 'companyName', 'jobUploaded']])
```
